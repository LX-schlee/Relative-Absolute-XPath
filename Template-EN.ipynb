```python
from selenium import webdriver
```

```python
# store website inside a variable
website = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
```

### Relative XPath

```python
# initialize Chrome Driver
driver = webdriver.Chrome('C:/webdrivers/chromedriver.exe')

# open website
driver.get(website)

# maximize window
driver.maximize_window()

countries = driver.find_elements_by_xpath('//td[@scope="row"]/a')

for country in countries:
    print(country.text)
```

### Absolute XPath

```python
# initialize Chrome Driver
driver = webdriver.Chrome('C:/webdrivers/chromedriver.exe')

# open website
driver.get(website)

# maximize window
driver.maximize_window()

countries = driver.find_elements_by_xpath('html/body/div[3]/div[3]/div[5]/div[1]/table/tbody//td[@scope="row"]/a')

for country in countries:
    print(country.text)
    
driver.close()
```
